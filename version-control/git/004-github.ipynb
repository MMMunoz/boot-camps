# Forking

### To interact with some else's code, the first step is to make a fork on Github, by clicking on the `fork` button
### This will create a copy of the repo that you can designate as your "`origin`"

### Clone your fork onto your machine: 

`$ git clone https://github.com/YOU/REPO_NAME.git`

(Note the `http*s*`)

## This makes your fork into a remote called `origin`. For example, try: 

`$ git remote`

## You probably want to immediately designate the original fork as a special remote, which we will call `upstream`

`$ git remote add upstream https://github.com/SOMEONE_ELSE/REPO_NAME.git`
    

## Take a look again at: 

`$ git remote -v`

The `-v` flag means *verbose*. That is, it will give you more information about your remotes

# Fetching

## This will bring in the changes to the code from the remote: 

`$ git fetch upstream`

# Merging

## And this will merge them in: 

`$git merge upstream/master`

# Alternatively: `git pull`

# Merging and conflict resolution

# Pushing your code 

## And getting it merged back upstream

`$ git branch new_branch master`

`$ git checkout new_branch`
